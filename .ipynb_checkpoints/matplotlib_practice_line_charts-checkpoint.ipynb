```python
%matplotlib inline
```

```python
#Import Dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot - using MATLAB interface
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt() - using object oriented interface
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#Anotate charts using Matlab method

# Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# Set the y limit between 0 and 50.
plt.ylim(0, 50)

# Create a title.
plt.title("PyBer Fare by Month")

# Add the legend.
plt.legend()
```

```python
# Using MatLab method

#Setting the figutre size
#plt.figure(figsize=(20,10))
plt.subplots(figsize=(10,10))

# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="red", linewidth=2, label='Boston')

# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")

# set the x axis limit between Jan and Jun
plt.xlim(0, 5)

# Set the y limit between 0 and 50.
plt.ylim(0, 50)

# Create a title.
plt.title("PyBer Fare by Month")

# Add a grid.
plt.grid()

# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt() - using object oriented interface

fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x_axis, y_axis,marker="*",color="green", linewidth=2)
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Month")
ax.set_ylabel("Fares ($)")
ax.legend("Boston")
ax.set_ylim(0,45)
ax.grid()
```
