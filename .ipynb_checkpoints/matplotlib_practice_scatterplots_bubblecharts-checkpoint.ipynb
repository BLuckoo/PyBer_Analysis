```python
#Import Dependencies
import matplotlib.pyplot as plt
```

## Create Scatter plot using MATLAB Method

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis, 'o')
# the lowercase "o" is for the scatter plot
```

```python
plt.scatter(y_axis,x_axis, c="red")
plt.legend("Chicago")
plt.title("PyBer Fares by Month")
plt.xlabel("Fares($)")
plt.ylabel("Month")
plt.gca().invert_yaxis()
plt.xlim(0,50)
```

## Bubble Chart using Matlab Method

```python
plt.scatter(x_axis, y_axis, s=y_axis)
# adding the s above is to have the size of the plots change according to the value of the y axis
```

```python
# in this example we are making the data points 3 times larger than the values of the y-axis

y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# using Python's list comprehension to replace the for loop above
plt.scatter(x_axis, y_axis, s = [i * 3 for i in y_axis])
```

## Scatter plot using Object Oriented interface

```python
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

## Bubble Chart using Object Oriented interface

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, c="skyblue",s = [i * 5 for i in y_axis], alpha = .5, edgecolor = 'black', linewidth=2 )
ax.legend("Boston")
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Month")
ax.set_xlim(0,50)
ax.invert_yaxis()
```
