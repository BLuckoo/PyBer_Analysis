```python
#Import Dependencies
import matplotlib.pyplot as plt
```

## Bar charts using the MatLab Method

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot - vertical bar chart
plt.bar(x_axis, y_axis)
```

```python
# Create the plot - anotate the bar chart
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot - horizontal bar chart - used with nominal variables
plt.barh(x_axis, y_axis)
```

```python
# Create the plot. - using the gco method with the invert method to sort they-axis in descending order
plt.barh(x_axis, y_axis)
plt.gca().invert_yaxis()
```

```python
# Create the plot. - using the gco method with the invert method to sort they-axis in descending order
plt.barh(x_axis, y_axis, color="magenta",label="Boston")
plt.legend()
plt.title("PyBer Fare by Month")
plt.xlabel("Fares ($)")
plt.ylabel("Month")
plt.gca().invert_yaxis()
```

## Bar Charts using the Object-Oriented approach

```python
# Create the plot with ax.plt() - Vertical bar chart
fig, ax = plt.subplots(figsize=(10,5))
ax.bar(x_axis, y_axis,color="black")
ax.legend("New York")
```

```python
# Create the plot with ax.plt() - Horizontal bar chart
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan")
ax.legend("Chicago")
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Month")
ax.invert_yaxis()
```
